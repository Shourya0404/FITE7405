```python
from AmericanOption import American
from BasketOption import Basket
from AsianOption import Asian
title= 'Arithmetic Asian'
```

### How Strike Price Affects Call Option Price

```python
S = 50
t = 0
T = 0.5
sigma = 0.2
r = 0.01
n = 50
M = 10**5
price=[]
param=[]
for K in range(10,100,10):
    C=Asian(S,K,T,sigma,r,n,'call',M ,True)
    price.append(C.arithmetic_asian()[0])
    param.append(K)

plt.plot(param,price)
# naming the x axis
plt.xlabel('Strike Price')
# naming the y axis
plt.ylabel('Call Option Price')
plt.title(title+' Option')
```

### How Volatility Affects Call Option Price

```python
fig, ax = plt.subplots(3, sharex=True, figsize=[11, 9])

fig.suptitle(title+ 'Option')
fig.align_labels()

S = 50
K= 35
t = 0
T=0.5
r = 0.01
n = 50
M = 10**5
price=[]
param=[]
for sigma in range(1,100,1):
    sigma=sigma*0.01
    C=Asian(S,K,T,sigma,r,n,'call',M, True)
    price.append(C.arithmetic_asian()[0])
    param.append(sigma)
ax[0].plot(param,price, label = 'ITM', color='r')
ax[0].set_ylabel('Call Option price')

S = 50
K= 50
t = 0
T=0.5
r = 0.01
n = 50
M = 10**5
price=[]
param=[]
for sigma in range(1,100,1):
    sigma=sigma*0.01
    C=Asian(S,K,T,sigma,r,n,'call',M, True)
    price.append(C.arithmetic_asian()[0])
    param.append(sigma)
ax[1].plot(param,price, label = 'ATM', color='b')
ax[1].set_ylabel('Call Option price')


S = 50
K= 60
t = 0
T=0.5
r = 0.01
n = 50
M = 10**5
price=[]
param=[]
for sigma in range(1,100,1):
    sigma=sigma*0.01
    C=Asian(S,K,T,sigma,r,n,'call',M, True)
    price.append(C.arithmetic_asian()[0])
    param.append(sigma)
ax[2].plot(param,price, label = 'OTM', color='g')
ax[2].set_ylabel('Call Option price')


fig.legend(loc='upper left')
plt.xlabel('Volatility')
```
