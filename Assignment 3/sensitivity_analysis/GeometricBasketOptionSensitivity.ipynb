### How Strike Price Affects Call Option Price

```python
S = [50, 50]
t = 0
T = 0.5
sigma = [0.3, 0.3]
r = 0.01
rho = 0.5
price=[]
param=[]
for K in range(10,100,10):
    C=Basket(S,K,T,sigma,r,rho,'call')
    price.append(C.geometric_basket())
    param.append(K)
plt.plot(param,price)
# naming the x axis
plt.xlabel('Strike Price')
# naming the y axis
plt.ylabel('Call Option Price')
plt.title(title+' Option')
```

### How Volatility Ratio between 2 Assets affects Call Price

```python
fig, ax = plt.subplots(3, sharex=True, figsize=[11, 9])

fig.suptitle(title+ ' Option')
fig.align_labels()

S = [50,50]
K= 40
t = 0
T=0.5
r = 0.01
sigma1 = 0
sigma2 = 1
rho = 0.5
price=[]
param=[]
for sigma in range(1,90,1):
    sigma1+=0.01
    sigma2-=0.01
    C=Basket(S,K,T,[sigma1,sigma2],r,rho,'call')
    price.append(C.geometric_basket())
    param.append(sigma1/sigma2)

ax[0].plot(param,price, label = 'ITM', color='r')
ax[0].set_ylabel('Call Option price')

S = [50,50]
K= 50
t = 0
T=0.5
r = 0.01
sigma1 = 0
sigma2 = 1
rho = 0.5
price=[]
param=[]
for sigma in range(1,90,1):
    sigma1+=0.01
    sigma2-=0.01
    C=Basket(S,K,T,[sigma1,sigma2],r,rho,'call')
    price.append(C.geometric_basket())
    param.append(sigma1/sigma2)

ax[1].plot(param,price, label = 'ATM', color='b')
ax[1].set_ylabel('Call Option price')

S = [50,50]
K= 60
t = 0
T=0.5
r = 0.01
sigma1 = 0
sigma2 = 1
rho = 0.5
price=[]
param=[]
for sigma in range(1,90,1):
    sigma1+=0.01
    sigma2-=0.01
    C=Basket(S,K,T,[sigma1,sigma2],r,rho,'call')
    price.append(C.geometric_basket())
    param.append(sigma1/sigma2)

ax[2].plot(param,price, label = 'OTM', color='g')
ax[2].set_ylabel('Call Option price')

fig.legend(loc='upper left')
plt.xlabel('Volatility Ratio')
```

### How Spot Price Ratio between 2 Assets affects Call Price

```python
S1 = 1 
S2 = 100
K= 50
t = 0
T=0.5
r = 0.01
sigma = [0.3,0.3]
rho = 0.5
price=[]
param=[]
for S in range(1,96,1):
    S1+=1
    S2-=1
    C=Basket([S1,S2],K,T,sigma,r,rho,'call')
    price.append(C.geometric_basket())
    param.append(S1/S2)
plt.plot(param,price)
# naming the x axis
plt.xlabel('Spot Price Ratio')
# naming the y axis
plt.ylabel('Call Option Price')
plt.title(title+ ' Option')
```
